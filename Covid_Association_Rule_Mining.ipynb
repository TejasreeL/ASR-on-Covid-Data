```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules

import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
data = pd.read_csv('Data LIWC 01 02 23.csv')
```

```python
data.dtypes
```

#Binning continuous data

```python
bins_info = {}
# continuous_columns = [
#     "retweet_count", "reply_count", "like_count", "quote_count", "Buzz",
#     "Clout", "Cognition", "Affect", "emo_pos", "emo_neg",
#     "we", "tentat", "Drives", "i", "we.1", "they", "insight", "cause",
#     "discrep", "certitude", "Positive", "Negative"
# ]
continuous_columns = [
    "retweet_count", "reply_count", "like_count", "quote_count",
    "Buzz", "emo_pos", "emo_neg", "Positive", "Negative"
]
binned = pd.DataFrame()

for col in continuous_columns:
    if col in data.columns:
        bins = np.linspace(data[col].min(), data[col].max(), 6)
        labels = [1, 2, 3, 4, 5]
        binned[col] = pd.cut(data[col], bins=bins, labels=labels, include_lowest=True).astype(float)
        bins_info[col] = {
            "min": data[col].min(),
            "max": data[col].max(),
            "bins": bins.tolist(),
            "labels": labels
        }
binned
```

```python
for col, info in bins_info.items():
    print(f"Column: {col}")
    for i in range(len(info["bins"]) - 1):
        print(f"  {info['labels'][i]}: {info['bins'][i]} to {info['bins'][i+1]}")
    print()
```

```python
columns_to_use = continuous_columns
subset = binned[columns_to_use].copy()

for column in columns_to_use:
    subset[column] = subset[column].astype('category')
```

```python
binned_encoded = pd.get_dummies(subset)
binned_encoded
```

```python
binned_encoded.dtypes
```

#Encoding categorical data

```python
columns_to_use = ['State', '3-Phase', 'Vividness', 'ContentType', 'DateDay', 'Time', 'Positive']
subset = data[columns_to_use].copy()

for column in columns_to_use:
    subset[column] = subset[column].astype('category')
```

```python
encoded = pd.get_dummies(subset)
encoded
```

#Combining encoded data with binned data

```python
combined = pd.concat([binned_encoded, encoded], axis=1)
combined
```

#Apriori for encoded data



```python
frequent_itemsets = apriori(encoded, min_support=0.4, use_colnames=True)
frequent_itemsets
```

```python
confidence_levels = [0.7]
results = {}

for confidence in confidence_levels:
    rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=confidence, num_itemsets=10)
    results[confidence] = rules
    print(f"\nRules for confidence level {confidence}:\n", rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])

for confidence, rules in results.items():
    print(f"Patterns for Confidence Level {confidence * 100}%:")
    if rules.empty:
        print("No rules found.\n")
    else:
        for _, row in rules.iterrows():
            antecedent = ', '.join(list(row['antecedents']))
            consequent = ', '.join(list(row['consequents']))
            print(f"If {antecedent}, then {consequent} (Support: {row['support']:.2f}, Confidence: {row['confidence']:.2f}, Lift: {row['lift']:.2f})")
        print("\n")
```
